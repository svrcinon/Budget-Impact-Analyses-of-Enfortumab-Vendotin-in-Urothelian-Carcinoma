Data generation for Power BI BIM (Budget Impact Model)


```python
import pandas as pd
import numpy as np
import math

# Generate 10,000 samples 
sample = 10000
patient_id = np.arange(1,sample+1,dtype=int) # 1st COLUMN


# Weight is Czech average (number are in kg and from UZIS page)
weight_man = 83.6
weight_woman = 69.2

# varience is not known, data are estimated
var_man = 14**2
var_woman = 11.5**2


# for lognormal function 
mu_man = np.log(weight_man)
mu_woman= np.log(weight_woman)
 
sigma_man = np.sqrt(np.log(1 + var_man / weight_man**2))
sigma_woman = np.sqrt(np.log(1 + var_woman / weight_woman**2))

# print(mu_man)
# print(mu_woman)
# print(sigma_man)
# print(sigma_woman)


# Does the urothelial cancer happend more to woman or men?
# https://www.rockymountaincancercenters.com/blog/bladder-cancer-in-women-versus-men-whats-the-difference
#the likelihood of men developing bladder cancer in their lifetime is about 1 in 27, compared to 1 in 89 for women
probability_of_men = 0.7672

rng = np.random.default_rng(42)
is_man = rng.random(sample) < probability_of_men


sex = np.where(
    is_man,
    "Male",
    "Female",
)# 2nd COLUMN


weights = np.where(
    is_man,
    np.round(rng.lognormal(mean=mu_man,   sigma=sigma_man,   size=sample),2),
    np.round(rng.lognormal(mean=mu_woman, sigma=sigma_woman, size=sample),2),
)# 3rd COLUMN


# print(len(weights))
# print(weights)

# dose that was given to the patient
dispense_dose = np.minimum(
   weights * 1.25,
    125
)

# dose that was charged = dose that were given to the patient + waste
charged_dose = np.minimum(
    np.ceil((weights * 1.25) / 10) * 10,
    130
)

mean_of_required_dose = np.mean(dispense_dose)
#print(mean_of_required_dose)

waste = charged_dose -dispense_dose
```

```python
columns = {"Id": patient_id , "Weight": weights,"Sex": sex ,"Dispense dose": dispense_dose ,"Charged dose": charged_dose ,"Waste": waste}
```

```python
df_patients = pd.DataFrame(data = columns)
```
